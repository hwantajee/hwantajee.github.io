## 암검진 (양성/악성 구분) with DecisionTreeClassifier, RandomForest

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
# print(cancer)
```

```python
data = cancer.data
label = cancer.target
print(data.shape)       
print(label.shape) 
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_label, test_label = train_test_split(data, label, test_size=0.3, random_state=1)
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(max_depth=5, random_state=1)
model = dtc.fit(train_data, train_label)

print(model.score(train_data, train_label))
print(model.score(test_data, test_label))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
predicts = model.predict(test_data)
print(confusion_matrix(test_label, predicts))
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_label, predicts))
```

```python
from sklearn.metrics import roc_auc_score
print(roc_auc_score(test_label, predicts))
# 큰 숫자일수록 좋음
```

```python
malignant = cancer.data[cancer.target == 0]
benign = cancer.data[cancer.target == 1]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (15, 10))
for i in range(len(cancer.feature_names)):
    plt.subplot(5, 6, i+1)
    plt.hist(malignant[:, i], bins=20, label="malignanat", alpha=0.5)
    plt.hist(benign[:, i], bins=20, label="benign", alpha=0.5)
    plt.title(cancer.feature_names[i])
plt.legend()
plt.show()
```

```python
from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz
from IPython.display import Image

dot_data = export_graphviz(dtc, out_file=None, feature_names=cancer.feature_names, \
                           class_names=cancer.target_names, filled=True, rounded=True, special_characters=True)
graph = graph_from_dot_data(dot_data)
# graph.write_png("cancer.png")
Image("./cancer.png")
```

```python
import seaborn as sns
import pandas as pd

df = pd.DataFrame(data, columns=cancer.feature_names)
sy = pd.Series(label, dtype="category")
sy = sy.cat.rename_categories(cancer.target_names)
df["class"] = sy
sns.pairplot(vars=["worst radius", "worst texture", "worst perimeter", "worst area"], hue="class", data=df)
plt.show()
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier(n_estimators=10, random_state=1)
#n_estimators(나무 갯수)
model = forest.fit(train_data, train_label)

print(model.score(train_data, train_label))      
print(model.score(test_data, test_label))          
```

```python
n_features = cancer.data.shape[1]           
# 변수 30개

plt.rcParams["figure.figsize"] = (10, 10)
plt.barh(range(n_features), model.feature_importances_, align="center")
plt.yticks(np.arange(n_features), cancer.feature_names)
plt.xlabel("importances")
plt.ylabel("features")
plt.show()
```

#### 최적화

```python
from sklearn.model_selection._search import GridSearchCV
```

```python
params = {"n_estimators": [5, 10, 15, 20], 
          "max_depth": [4, 5, 6],
          "min_samples_leaf": [10, 12, 14],
          "min_samples_split": [10, 12, 14]}
```

```python
cv = GridSearchCV(RandomForestClassifier(random_state=1), params, n_jobs=1)          
# n_jobs(cpu 작업 갯수; 대게 1)
```

```python
models = cv.fit(train_data, train_label)
model = models.best_estimator_
print(models.best_params_)
```

```python
result_df = pd.DataFrame(models.cv_results_)
print(result_df)
```

```python
result_df = result_df[["params", "mean_test_score", "rank_test_score"]]
print(result_df)
```

```python
from sklearn import metrics
```

```python
predict = model.predict(test_data)
report = metrics.classification_report(test_label, predict)
print(report)
```
